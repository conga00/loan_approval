## Handle duplicated rows

```python
detect_dup_series = raw_data.duplicated(keep='first')
num_duplicated_rows = detect_dup_series.sum()
if num_duplicated_rows == 0:
    print(f"Raw data have no duplicated rows!")
else:
    if num_duplicated_rows > 1:
        ext = "lines"
    else:
        ext = "line"
    print(f"Raw data have {num_duplicated_rows} duplicated " + ext + ". Please de-deduplicate the data!")
```

## Inspect and handle columns name

```python
raw_data.columns.values
```

Many columns' name have a space before their name. We can fix this by mapping with new names which do not contain spaces

```python
new_cols_name = ['loan_id', 'no_of_dependents', 'education', 'self_employed',
                 'income_annum', 'loan_amount', 'loan_term', 'cibil_score',
                 'residential_assets_value', 'commercial_assets_value',
                 'luxury_assets_value', 'bank_asset_value', 'loan_status']

column_name_mapping = dict(zip(raw_data.columns, new_cols_name))
raw_data.rename(columns=column_name_mapping, inplace=True)

raw_data.columns.values
```

## Removing meaningless columns

```python
processed_data = raw_data.drop(['loan_id'],axis= 1)
```

### Numeric data inspection

```python
num_data = processed_data.select_dtypes(exclude='object')
def missing_ratio(df):
    return (df.isna().mean()).round(3)

def median(df):
    return (df.quantile(0.5)).round(3)

def lower_quartile(df):
    return (df.quantile(0.25)).round(3)

def upper_quartile(df):
    return (df.quantile(0.75)).round(3)

num_data_info = num_data.agg([missing_ratio, "min", lower_quartile, median, upper_quartile, "max"])
num_data_info
```

```python
sns.set()
num_data.hist(figsize=(20,12),bins =32)
plt.show()
```

**Observation:** most of columns were pretty balance as uniform distribution appeared in many graphs. 'residential_assets_value' and 'commercial_assets_value' are exponential distribution. 'bank_asset_value', 'luxury_assets_value' and 'loan_amount' seemed like poisson distribution pdf. These columns might need some mathematical transformation if needed.

### Categorical data inspection

```python
cat_data = processed_data.select_dtypes(include='object')

def missing_ratio(s):
    return (s.isna().mean() * 100).round(3)

def num_values(s):
    s = s.str.split(';')
    s = s.explode()
    return len(s.value_counts())

def value_ratios(s):
    s = s.str.split(';')
    s = s.explode()
    totalCount = (~s.isna()).sum()
    return ((s.value_counts()/totalCount*100).round(3)).to_dict()

cat_data_info = cat_data.agg([missing_ratio, num_values, value_ratios])
cat_data_info
```

```python
cat_data['education'].value_counts()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 10))
for ax, col in zip(axes, cat_data.columns):
    data = cat_data[col].value_counts()
    ax.pie(data, autopct=lambda p: f'{p:.2f}%', labels=data.index, startangle=30)
    ax.set_title(f'Pie chart of {col}')

plt.show()
```

**Number of approved in loan status is higher than rejected. On the other hand, other categorical columns are balanced**

# 3. Data exploring

I will attempt to visualize the relationship between target variable (loan_status) and independent features.

```python
eda_df = processed_data.copy()
```

## Education

```python
edu_df = eda_df.copy()

edu_df = edu_df.groupby(['education','loan_status']).size().unstack()
edu_df.plot(kind='bar', stacked=True,color=['lightgreen', 'lightcoral'], figsize=(10, 6),)

plt.title('Loan Status by Education Level')
plt.xlabel('Education Level')
plt.ylabel('Count')
plt.show()
```

**It seemed graduated or not graduated did not determine loan status much as rejection ratio in these two groups is approximately similar**

## Self Employed

```python
empl_df = eda_df.copy()

empl_df = empl_df.groupby(['self_employed','loan_status']).size().unstack()
empl_df.plot(kind='bar', stacked=True,color=['lightgreen', 'lightcoral'], figsize=(10, 6),)

plt.title('Loan Status by self employment')
plt.xlabel('Self employment')
plt.ylabel('Count')
plt.show()
```

**The case is similar to Education scenerio as loan rejection ratio in two categories of self employment status are not very different**

## Number of dependence

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='no_of_dependents', data=eda_df, hue='loan_status', edgecolor=".3", linewidth=.5)

plt.title('Distribution of Number of Dependents by Loan Status')
plt.xlabel('Number of Dependents')
plt.ylabel('Count')
plt.show()
```

**As number of dependents increased, rejection ratio increased slightly, howvever, not very noticable.**

**Function that helps to annotate rejection/approval ratio for each bins**

```python
def hist_annotation(histplot):

    bin_count = [0 for i in range(len(histplot.containers[0]))]
    
    for container in histplot.containers:
        bin_count = bin_count + container.datavalues
    
    cumulative_height = [0 for _ in range(len(bin_count))]
    for container in histplot.containers:
        for i,p in enumerate(container.patches): 
            cumulative_height[i] = cumulative_height[i] + p.get_height()
            histplot.annotate(f"{(p.get_height()/bin_count[i]):.1%}\n", (p.get_x() + p.get_width() / 2, cumulative_height[i]*0.9),
                            ha="center", va="center", color= 'black', fontweight="bold",fontsize = 15)
```

## Annual income and loan amount

```python
fig, ax = plt.subplots(2,figsize = (25,15))

h1 = sns.histplot(x='income_annum', data=eda_df, hue='loan_status', multiple="stack", bins = 10,
             edgecolor=".3", linewidth=.5, ax=ax[0])

ax[0].set_title('Distribution of Annual Income by Loan Status',fontsize = 20)
ax[0].set_xlabel('Annual Income')
ax[0].set_ylabel('Count')

h2 = sns.histplot(x='loan_amount', data=eda_df, hue='loan_status', multiple="stack", bins = 10,
             edgecolor=".3", linewidth=.5, ax=ax[1])

ax[1].set_title('Distribution of Loan Amount by Loan Status', fontsize = 20)
ax[1].set_xlabel('Loan Amount')
ax[1].set_ylabel('Count')

hist_annotation(h1)
hist_annotation(h2)

plt.tight_layout()
plt.show()
```

- **In annual income, no real relationship shown as the rejection/approval ratio fluctuated when the x axis value increased**
- **Rejection ratio of loan amount increased as the money increased, although at one point, it stopped and started to decrease.**
